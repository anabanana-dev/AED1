```python
import random

n = random.randint(1, 50)
tentativas = 0
limite = 5

numero_escolhido = int(input("Digite um número entre 1 e 50: "))

while n != numero_escolhido and tentativas < limite:
    tentativas += 1                          
    restantes = limite - tentativas          

    if n > numero_escolhido:
        print("O número secreto é MAIOR, tente novamente.")
    else:
        print("O número secreto é MENOR, tente novamente.")

    if restantes > 0:                       
        print("Tentativas restantes" , restantes)
        numero_escolhido = int(input("Digite um número entre 1 e 50: "))


if n == numero_escolhido:
    print("Acertou em", tentativas + 1, "tentativa(s)! Parabéns!!")
else:
    print("Suas", limite, "tentativas acabaram! O número secreto era", n)
```

```python
import random

n = random.randint(1, 50)
acertou = False

for tentativa in range(10):
    num = int(input("Tentativa " + str(tentativa + 1) + "/10 - Digite um numero: "))
    if num == n:
        acertou = True
        break
    elif n > num:
        print("O numero secreto e MAIOR! Restam " + str(9 - tentativa) + " tentativas.")
    else:
        print("O numero secreto e MENOR! Restam " + str(9 - tentativa) + " tentativas.")

if acertou:
    print("Parabens! Acertou na tentativa " + str(tentativa + 1) + "!")
else:
    print("Suas tentativas acabaram! O numero secreto era " + str(n) + ".")
```

```python
import random

jogar_novamente = "s"

while jogar_novamente == "s":

    n = random.randint(1, 50)
    acertou = False
    tentativas = 0

    while tentativas < 10 and not acertou:
        tentativas += 1
        num = int(input("Tentativa " + str(tentativas) + "/10 - Digite um numero: "))

        diferenca = abs(num - n)

        if num == n:
            acertou = True
        elif diferenca <= 2:
            print("Quente! Muito perto!")
        elif diferenca <= 10:
            print("Morno...")
        else:
            print("Frio! Muito longe.")

        restantes = 10 - tentativas
        if not acertou and restantes > 0:
            print("Tentativas restantes: " + str(restantes))

    if acertou:
        print("Parabens! Acertou na tentativa " + str(tentativas) + "!")
    else:
        print("Suas tentativas acabaram! O numero secreto era " + str(n) + ".")

    jogar_novamente = input("Deseja jogar de novo? (s/n): ")

print("Obrigado por jogar!")
```
