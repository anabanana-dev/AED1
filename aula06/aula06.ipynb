```python
nota1 = float(input("Qual a primeira nota do Aluno?"))
if nota1 >= 7:
        nota1 = nota1 + 0.5
nota2 = float(input("Qual a segunda nota do Aluno?"))
if nota2 >= 7:
        nota2 = nota2 + 0.5
media = (nota1 + nota2)/2
print("A média é igual a : ", media)
if media >= 5:
    print("Aprovado!")
else:
    print("Reprovado")

```

```python
ladoE = float(input("Digite o valor do lado Esquerdo"))
ladoD = float(input("Digite o valor do lado Direito"))
base = float(input("Digite o valor da base"))

if ladoE + ladoD >= base:
    print("isso é um triângulo")
    if (ladoE == ladoD) and (ladoE == base) and (ladoD == base):
        print("é um triangulo Equilatero")
    elif (ladoE == ladoD) or (ladoD == base) or (ladoE == base):
        print("é um triangulo isosceles")
    else:
        print("é um triangulo escaleno")
else:
    print("Não é um triangulo")
    
```
