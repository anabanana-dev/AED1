```python
numero = int(input("Digite um número inteiro: "))

if numero % 5 == 0 and numero % 3 == 0:
    print("O número" , numero, "é divisível por 5 e 3 ao mesmo tempo.")
else:
    print("O número" ,numero , "não atende aos dois critérios simultaneamente.")
```

```python
valor_compra = float(input("Digite o valor da compra: R$ "))

if valor_compra <= 1000:
    percentual = 0.10 
elif valor_compra <= 5000:
    percentual = 0.20  
else:
    percentual = 0.30 

valor_desconto = valor_compra * percentual
valor_final = valor_compra - valor_desconto

print("\n--- Resumo da Compra ---")
print("Valor original: R$", valor_compra)
print("Desconto aplicado:", percentual * 100, "%")
print("Valor do desconto: R$", valor_desconto)
print("Valor total a pagar: R$", valor_final)
```

```python
largura = float(input("Digite a largura do aposento (m): "))
comprimento = float(input("Digite o comprimento do aposento (m): "))
litros_lata = float(input("Digite quantos litros tem a lata de tinta: "))

altura_pe_direito = 2.80
area_porta = 0.80 * 2.10
rendimento_por_litro = 3 
area_total_paredes = (2 * largura * altura_pe_direito) + (2 * comprimento * altura_pe_direito)

area_a_pintar = area_total_paredes - area_porta

total_litros_necessarios = area_a_pintar / rendimento_por_litro

quantidade_latas = total_litros_necessarios / litros_lata

print("\n--- Relatório de Pintura ---")
print("Área total a pintar: ", area_a_pintar, "m²")
print("Total de litros necessários: ", total_litros_necessarios, "litros")
print("Quantidade de latas de tinta: ", quantidade_latas)
```
