### 3 -  FaĂ§a um programa que leia e valide as seguintes informaĂ§Ăľes: 
##### a. Nome: maior que 3 caracteres; 
##### b. Idade: entre 0 e 150; 
##### c. SalĂĄrio: maior que zero; 
##### d. Sexo: 'f' ou 'm'; 
##### e. Estado Civil: 's', 'c', 'v', 'd'; 

```python
while True:
    nome = input("Digite seu nome: ")
    if len(nome) <= 3:
        print("Erro: O nome deve ter mais de 3 letras. Vamos recomeĂ§ar!\n")
        continue

    idade = int(input("Digite sua idade: "))
    if idade <= 0 or idade > 150:
        print("Erro: A idade deve estar entre 1 e 150 anos. Vamos recomeĂ§ar!\n")
        continue


    salario = float(input("Digite seu salario: "))
    if salario <= 0:
        print("Erro: O salĂĄrio deve ser maior que zero. Vamos recomeĂ§ar!\n")
        continue


    sexo = input("Digite seu sexo (f ou m): ").lower()
    if sexo not in ['f', 'm']:
        print("Erro: O sexo deve ser apenas 'f' ou 'm'. Vamos recomeĂ§ar!\n")
        continue


    estado_civil = input("Digite seu estado civil (s, c, v ou d): ").lower()
    if estado_civil not in ['s', 'c', 'v', 'd']:
        print("Erro: O estado civil deve ser 's', 'c', 'v' ou 'd'. Vamos recomeĂ§ar!\n")
        continue

    print("Suas informaĂ§Ăľes foram cadastradas com sucesso!")
    break
```

### 8 - FaĂ§a um programa que calcule o mostre a mĂŠdia aritmĂŠtica de N notas.

```python
n = int(input("Quantas notas vocĂŞ quer calcular? "))

soma_das_notas = 0

for i in range(n):

    nota = float(input("Digite a " + str(i + 1) + "ÂŞ nota: "))

    soma_das_notas += nota

media = soma_das_notas / n

print("A mĂŠdia aritmĂŠtica das", n, "notas ĂŠ:", round(media, 2))
```

### 10 - O Departamento Estadual de Meteorologia lhe contratou para desenvolver um programa que leia as um conjunto indeterminado de temperaturas, e informe ao final a menor e a maior temperaturas informadas, bem como a mĂŠdia das temperaturas.

```python
t = int(input("Quantas temperaturas vocĂŞ quer digitar? "))

soma = 0
maior = 0
menor = 0

for i in range(t):
    temperatura = float(input("Digite a " + str(i + 1) + "ÂŞ temperatura: "))

    soma += temperatura

    if i == 0:
        maior = temperatura
        menor = temperatura
    else:
        if temperatura > maior:
            maior = temperatura

        if temperatura < menor:
            menor = temperatura

if t > 0:
    media = soma / t
    print("A maior temperatura foi:", maior)
    print("A menor temperatura foi:", menor)
    print("A mĂŠdia das", t, "temperaturas ĂŠ:", round(media, 2))
else:
    print("Quantidade invĂĄlida.")
```
