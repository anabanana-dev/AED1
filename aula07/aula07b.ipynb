```python
n = int(input("Digite qntd de pessoas: "))
soma_peso = 0
soma_altura = 0
imc_maior = 0
imc_menor = 1000
x = 1

while x <= n:
    peso = float(input("Digite o peso em Kg: "))
    altura = float(input("Digite  a altura, metros: "))
    imc = peso / (altura ** 2)
    if imc < imc_menor:
        imc_menor = imc
    if imc > imc_maior:
        imc_maior = imc
    soma_peso += peso
    soma_altura += altura
    x += 1
media_peso = soma_peso / n
media_altura = soma_altura / n
print("imc maior = ", imc_maior, "imc menor = ", imc_menor)
print("Peso médio =  ", media_peso)
print("Altura média = ", media_altura)
```
